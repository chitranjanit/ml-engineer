```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,2*np.pi,0.01)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,label='Sin')
plt.plot(x,y2,label="Cos")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.title('Sine & Cosine Function')

plt.show()
```

```python
#It create a 3 dimensiional adarray
import numpy as np

a = np.arange(8).reshape(2,2,2)
print('The original array:')
print(a)
print('\n')

# now swap numbers between axis 0 ( along depth ) and axis 2 ( alon width )

print('The array after applying the swapaxes function:')
print( np.swapaxes(a,2,0) )
```
